### Line Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,12,8,15,7]
plt.plot(x,y,marker='o')
plt.title("Line plot")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Bar plot

```python
catge=['A','B','C','D']
values=[25,40,15,30]
colors = ['red', 'green', 'blue', 'orange']

plt.bar(catge,values,color=colors)
plt.title('Bar plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### Histogram

```python
data=[32,45,50,28,36,42,38,25,29,52,48,60]
plt.hist(data,bins=5,color="orange",edgecolor="black")
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency") 
plt.show()
```

### Scatter Plot

```python
x=[3,5,7,10,12,15]
y=[25,40,30,50,35,60]
plt.scatter(x,y,color="black",marker="o")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Pie Chart

```python
labels=["Apple","Banana","Oranges","Grapes"]
size=[30,25,20,15]
colors = ['lightblue', 'lightgreen', 'blue', 'orange']
plt.pie(size,labels=labels,autopct='%1.1f%%',colors=colors)
plt.show()
```

### Box plot

```python
data=[45,60,75,300,80,350,90,100,110,120,250,200]
plt.boxplot(data,vert=False)
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```

### Step Plot

```python
time=[1,2,3,4,5]
stock_prize=[100,110,105,120,130]
plt.step(time,stock_prize,color="orange",where="mid")
plt.title("Step Plot")
plt.xlabel("Time")
plt.ylabel("Stock Prize")
plt.show()
```
