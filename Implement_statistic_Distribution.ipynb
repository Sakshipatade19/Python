## Normal Distribution

### Question 1
**Suppose there are 100 students in the class and in one of the mathematics tests the average marks scored by the students in the subject is 78 and the standard deviation is 25. The marks of student follow Normal probability distribution.**
1. Percentage of student who got less than 60 marks.
2. Percentage of student who have scored more than 70 marks.

```python
from scipy.stats import norm
from scipy import stats
```

```python
U = 78
SD = 25
N= 100
score= 60
# Calculating the z-score
z_score= (score-U)/SD
prob = norm.cdf(z_score)
prob*100
```

- student who are getting score less than 60 are 23.57%

### Perform Annova Testing pract-6

```python
import numpy as np
method1=np.array([85 ,88 ,92 ,94 ,90])
method2=np.array([78 ,80 ,83 ,77 ,79])
method3=np.array([91 ,89 ,94 ,96 ,92])

f_stat , p_value=stats.f_oneway(method1,method2,method3)

print(f'F-statistic : {f_stat}')
print(f'P-Value : {p_value}')
```

p value is less than 0.05 that means there is a significant different between the 3 methods

### Two way Annova

```python
import pandas as pd
data = pd.DataFrame({
    'Score' :[85,78,91,88,80,89,92,83,94,94,77,96,90,79,92],
    'Method' : ['Method1','Method1','Method1','Method2','Method2','Method2','Method3','Method3',
                'Method3','Method1','Method2','Method3','Method1','Method2','Method3'],
    'Gender':['Female','Female','Female','Male','Male','Male','Female','Female','Female',
              'Male','Male','Male','Female','Female','Female']
})
```

```python
data
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

#fit the model
model=ols('Score ~ C(Method)*C(Gender)',data=data).fit()

# two way annova
annova_table=sm.stats.anova_lm(model,typ=2)
print(annova_table)

'''
C(Method) : test the effect of the method
C(Gender) : Test of the effect of Gender
C(Method):C(Gender) : Tests the interaction effect betwen
'''
```

- no significant differnce between the methods
- There is a significant difference between the gender
