```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.DataFrame(
    {
        'roll_no':[1,2,3,4,5],
        'name':['A','B','C','D','E'],
        'gender':['M','M','F','M','F'],
        'marks':['Good','Good','Poor','Average','excellent']

    }
)
```

## using pandas doing One hot encoding

```python
pd.get_dummies(data,columns=['gender','marks'],dtype=int)
```

## Using scikit-learn (skelarn) doing label encdoing

```python
le = LabelEncoder()
datanew=data
datanew['gender']=le.fit_transform(data['gender'])
datanew['marks']=le.fit_transform(data['marks'])
```
